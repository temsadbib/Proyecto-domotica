**Tareas 1 y 2:** correlaciones entre las **cuatro sondas** del aula (temperatura, humedad, presión) desde **`gold_correlaciones`** (`sql/05_gold_correlaciones.sql`).


```python
import pandas as pd
import psycopg2
import seaborn as sns
import matplotlib.pyplot as plt

conexion = psycopg2.connect(
    host="localhost",
    port="5432",
    database="postgres",
    user="postgres",
    password="Qwe1234."
)

query = """
SELECT
    temp_aula_1,
    temp_aula_2,
    temp_aula_3,
    temp_aula_4,
    temp_exterior
FROM gold_correlaciones;
"""

df_temperaturas = pd.read_sql_query(query, conexion)

correlaciones = df_temperaturas.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlación: cuatro temperaturas aula y exterior', fontsize=16)
plt.show()
```

```python
print("Datos registrados por sensor")
print(df_temperaturas.count())

print("\nHuecos vacíos por sensor")
print(df_temperaturas.isnull().sum())
```

```python
query = """
SELECT
    hum_aula_1,
    hum_aula_2,
    hum_aula_3,
    hum_aula_4
FROM gold_correlaciones;
"""

df_humedad = pd.read_sql_query(query, conexion)

correlaciones = df_humedad.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlación: cuatro humedades aula', fontsize=16)
plt.show()

query = """
SELECT
    pres_aula_1,
    pres_aula_2,
    pres_aula_3,
    pres_aula_4
FROM gold_correlaciones;
"""

df_presion = pd.read_sql_query(query, conexion)

correlaciones = df_presion.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlación: cuatro presiones aula', fontsize=16)
plt.show()
```

```python
print("Datos registrados por sensor")
print(df_humedad.count())

print("\nHuecos vacíos por sensor")
print(df_humedad.isnull().sum())

print("Datos registrados por sensor")
print(df_presion.count())

print("\nHuecos vacíos por sensor")
print(df_presion.isnull().sum())
```

```python
query_ventanas = """
SELECT
    min_puerta_1,
    min_ventana_1, min_ventana_2, min_ventana_3, min_ventana_4,
    min_ventana_5, min_ventana_6, min_ventana_7, min_ventana_8,
    min_ventana_9, min_ventana_10, min_ventana_11, min_ventana_12
FROM gold_features_horaria;
"""

df_ventanas = pd.read_sql_query(query_ventanas, conexion)

corr_ventanas = df_ventanas.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_ventanas, annot=True, cmap='viridis', fmt=".1f", linewidths=.5)
plt.title('Correlación: minutos abiertos puerta y ventanas (por hora)', fontsize=14)
plt.show()
```

```python
print("Datos registrados por sensor")
print(df_ventanas.count())

print("\nHuecos vacíos por sensor")
print(df_ventanas.isnull().sum())
```
